```python
basic_info_cols = [
    '_id', 'today', '_Exact Location_latitude', '_Exact Location_longitude',
    'Governorat', 'District', 'Is the company registered',
    'What is the Type of registration?', 'Business name:',
    'Economic sector:', 'Sub-sector:', 'Please specify the business products and services',
    'Phone Number', 'Business Phone Number (If applicable)',
    'Enterprise size in terms of number of employees:', 'Enterprise size (Micro, Small, Medium, Large)',
    'Do you have any current available vacancies?', 'Do you expect the need for employees in the next year?',
    'Number of expected job vacancies:', 'start_date', 'end_date', 'start_time', 'end_time'
]

# Columns related to current recruitment
current_recruitment_cols = [col for col in df.columns if col.startswith('Current//')]
current_recruitment_data = df[ + current_recruitment_cols].copy()
current_recruitment_data['Recruitment_Type'] = 'Current'

# Columns related to future recruitment
future_recruitment_cols = [col for col in df.columns if col.startswith('Future//')]
future_recruitment_data = df[df.columns+ future_recruitment_cols].copy()
future_recruitment_data['Recruitment_Type'] = 'Future'

# Renaming columns to remove prefixes for ease of use
current_recruitment_data.columns = [col.replace('Current//', '') for col in current_recruitment_data.columns]
future_recruitment_data.columns = [col.replace('Future//', '') for col in future_recruitment_data.columns]

# Combining current and future recruitment data
reshaped_df = pd.concat([current_recruitment_data, future_recruitment_data], ignore_index=True)
```
